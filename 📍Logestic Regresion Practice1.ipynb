```python
import pandas as pd
from matplotlib import pyplot as pyplot
%matplotlib inline
```

```python
df= pd.read_csv("/content/cardata.csv")
df.head(20)
```

```python
# @title Age vs EstimatedSalary

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Age', y='Purchased', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
plt.scatter(df.Age, df.Purchased, marker='+', color='red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['Age']], df.Purchased, train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(x_train, y_train)
```

```python
x_test
```

```python
y_predicted= model.predict(x_test)
y_predicted
```

```python
model.predict_proba(x_test)
```

```python
#confusion matrix

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predicted)
```

```python
model.score(x_test, y_test)
```
